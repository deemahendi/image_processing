 image segmentation using histogram thresholding works for separating the objects from the background when there are distinct local minima and maxima in the histograms. if we have  bi-modal histogram distribution  and there is one maximum for objects and one for background.  the two maxima are located relatively far away from each other and therefore the optimal threshold can be located somewhere in between the two peaks.  when the noise levels are high the method produces high level of pixel misclassification.
 

```python
import cv2 
import numpy as np 
from matplotlib import pyplot as plt 
img1 = cv2.imread('images/flower.pgm',0)
hist1 =cv2.calcHist([img1],[0],None,[256],[0,256])
plt.plot(hist1)
plt.show()

img = cv2.imread('images/flower.pgm')
grayscaled = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#retval, threshold = cv2.threshold(grayscaled, 200,700, cv2.THRESH_BINARY)
retval, threshold = cv2.threshold(grayscaled,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
adaptive = cv2.adaptiveThreshold(grayscaled,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,815,1)
titles=['Orginal Flower','gray','OTSU','ADAPTIVE']
images=[img1,img,threshold,adaptive]

for i in range(4):
    plt.subplot(2,4,i+1),plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
plt.show()
```
