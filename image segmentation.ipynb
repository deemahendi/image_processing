```python
import cv2    # include  OpenCV library in my code
import numpy as np # numpy library for array in python np is the object we use it with its function
from matplotlib import pyplot as plt

img = cv2.imread('images/try4.jpg',0)
#hist,bins = np.histogram(img.ravel(),256,[0,256]) 
hist = cv2.calcHist([img],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()


plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
import cv2
import numpy as np

img = cv2.imread('images/try4.jpg')
grayscaled = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
retval, threshold = cv2.threshold(grayscaled, 230,900, cv2.THRESH_BINARY)
cv2.imshow('original',img)
cv2.imshow('threshold',threshold)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

img = cv2.imread('images/try4.jpg')
grayscaled = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
retval, threshold = cv2.threshold(grayscaled, 165,1000, cv2.THRESH_BINARY)
mask_inv = cv2.bitwise_not(threshold)
cv2.imshow('threshold',threshold)
cv2.imshow('mask_inv',mask_inv)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img1 = cv2.imread('images/try4.jpg')
img2 = cv2.imread('images/thre2.png')
final = cv2.multiply(img1/255, img2/255)

blur = cv2.blur(final,(5,5))

plt.show()
cv2.imshow('blur',blur)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
